```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
cereal = cv2.imread('cereal.png')
cereales = cv2.imread('cereales.png')
plt.imshow (cereal)
```

```python
cereal = cv2.cvtColor(cereal, cv2.COLOR_BGR2RBG)
cereales = cv2.cvtColor(cereales, cv2.COLOR_BGR2RBG)
plt.imshow (cereal)
```
