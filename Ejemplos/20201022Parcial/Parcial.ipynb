```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
imagen = cv2.imread('GetPersonaPhoto.jpg')
```

```python
plt.imshow(imagen)
```

```python
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen)
```
