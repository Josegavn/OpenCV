```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
imagen = np.zeros((700,900,3), dtype=np.int16) 
```

```python
imagen.shape
```

```python
plt.imshow(imagen)
```

```python
cv2.rectangle(imagen, pt1=(36, 36), pt2=(236, 236), color= (14, 226, 237), thickness=10)
cv2.rectangle(imagen, pt1=(664, 36), pt2=(864, 226), color= (14, 226, 237), thickness=10) 
plt.imshow(imagen)
```

```python
cv2.circle(imagen, center=(450, 300), radius=80, color=(7, 234, 63), thickness=10)
plt.imshow(imagen)
```

```python
cv2.line(imagen, pt1=(0, 12), pt2=(900, 12), color=(237, 226, 14), thickness=5)
cv2.line(imagen, pt1=(12, 0), pt2=(12, 700), color=(237, 226, 14), thickness=5)
cv2.line(imagen, pt1=(0, 688), pt2=(900, 688), color=(237, 226, 14), thickness=5)
cv2.line(imagen, pt1=(888, 0), pt2=(888, 700), color=(237, 226, 14), thickness=5)
plt.imshow(imagen)
```

```python
cv2.ellipse(imagen,(200, 500),(100, 50),0,60,360,(237, 14, 62), (20))
plt.imshow(imagen)
```
