```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread('amigos.jpg')
```

```python
plt.imshow(imagen)
```

```python
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen)
```

```python
imagen_2 = cv2.cvtColor
```

```python
metodo = cv2.TM_CCOEFF
#metodo = cv2.TM_CCORR
#metodo = cv2.TM_SQDIFF
```

```python
resultado = cv2.matchTemplate(imagen, imagen_2, metodo)
```

```python
plt.imshow(resultado)
```

```python
valor_min, valor_max, pos_min, pos_max =cv2.minMaxLoc(resultado)
```

```python
alto, ancho, colores = imagen_2.shape
```

```python
#posiciones
top_izquierda = pos_max
#posición en X                  (y,x)
boton_derecha = (pos_max [0]+ancho, pos_max[1+alto])
```

```python
cv2.rectangle(imagen, top_izquierda, boton_derecha, (255,0,0), thickness= 8)
```
